#### Feature Wise Visualization

```python
## 1. Gender Column
```

```python
# How is distribution of Gender ?

fig,axs=plt.subplots(1,2,figsize=(8,4))
sns.countplot(data=df,x=df['gender'],ax=axs[0])
for container in axs[0].containers:
    axs[0].bar_label(container)

plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],autopct='%.2f%%')
plt.show()
```

##### Gender has apprx balanced data.

```python
# Is gender has any impact on student's performance ?

gender_group = df.groupby('gender').mean()
gender_group
```

```python
plt.figure(figsize=(8,6))

X = ['Total Average','Math Average']

female_scores = [gender_group['average score'][0],gender_group['math score'][0]]
male_scores = [gender_group['average score'][1],gender_group['math score'][1]]
x_axis = np.arange(len(X))
plt.bar(x_axis-0.2,male_scores,0.4,label='Male')
plt.bar(x_axis+0.2,female_scores,0.4,label='Female')

plt.xticks(x_axis,X)
plt.ylabel('Marks')
plt.legend()
plt.show()
```

##### In total, females have better score than males whereas Males scored higher in math. 

```python
## 2. Race/Ethnicity Column
```

```python
# How is group wise distribution ?

fig,axs=plt.subplots(1,2,figsize=(8,4))
sns.countplot(data=df,x=df['race/ethnicity'],ax=axs[0])
for container in axs[0].containers:
    axs[0].bar_label(container)

plt.pie(x=df['race/ethnicity'].value_counts(),labels=df['race/ethnicity'].value_counts().index,autopct='%.2f%%')
plt.show()
```

##### most of the students belonging from group C and group D.
##### group A has the lowest number of students

```python
# Is race/ethnicity has any impact on student's performance ?

race_group = df.groupby('race/ethnicity')

f,ax=plt.subplots(1,3,figsize=(20,8))
sns.barplot(x=race_group['math score'].mean().index,y=race_group['math score'].mean().values,ax=ax[0])
ax[0].set_title('Math Score')

for container in ax[0].containers:
    ax[0].bar_label(container)

sns.barplot(x=race_group['reading score'].mean().index,y=race_group['reading score'].mean().values,ax=ax[1])
ax[1].set_title('Reading Score')

for container in ax[1].containers:
    ax[1].bar_label(container)

sns.barplot(x=race_group['writing score'].mean().index,y=race_group['writing score'].mean().values,ax=ax[2])
ax[2].set_title('Writing Score')

for container in ax[2].containers:
    ax[2].bar_label(container)

plt.show()
```

##### group E students have scored highest marks while group A scored lowest.

```python
## 3. Parental Level of Education Column
```

```python
# What is educational background of student's parent ?

plt.figure(figsize=(13,7))
sns.countplot(data=df,x=df['parental level of education'])
plt.title('Comparison of Parental Education')
plt.xlabel('degree')
plt.ylabel('count')
plt.show()
```

##### largest number of parents are from 'some college'.

```python
# Is parental education has any impact on student's performance ?

df.groupby('parental level of education').agg('mean').plot(kind='barh')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

##### parents possess master's and bachelor's degree are higher than others.

```python
## 4. Lunch Column
```

```python
# Which type of lunch is most common among students ?

plt.figure(figsize=(4,3))
sns.countplot(data=df,x=df['lunch'])
plt.title('Comparison of types of Lunch')
plt.xlabel('types of lunch')
plt.ylabel('count')
plt.show()
```

##### students being served standard lunch was more than free/reduced lunch.

```python
# What is the effect of lunch type on test results ?

f,ax=plt.subplots(1,2,figsize=(20,8))
sns.countplot(x=df['parental level of education'],data=df,hue='test preparation course',ax=ax[0])
ax[0].set_title('Students vs Test Results')
for container in ax[0].containers:
    ax[0].bar_label(container)

sns.countplot(x=df['parental level of education'],data=df,hue='lunch',ax=ax[1])
for container in ax[1].containers:
    ax[0].bar_label(container)

plt.show()
```

##### students who get standard lunch perform better than free/reduced lunch.
